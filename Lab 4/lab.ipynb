```python
import torch
import random
import math

def quadratic_function(x):
    return x ** 2 - 2 * x + 30

def random_function(x):
    return torch.sin(x)+torch.cos(x**2+x)
x = torch.tensor([random.uniform(-10, 10)], requires_grad=True)
print("Starting x:", x.item())

for i in range(100):
    y = random_function(x)
    y.backward()
    with torch.no_grad():
        print(f"Grad: {x.grad}")
        x -= 0.01 * x.grad
    x.grad.zero_()
    print(f"Iteration {i + 1}: x = {x.item()}, f(x) = {y.item()}")
print(f"Optimized x: {x.item()}")
```
